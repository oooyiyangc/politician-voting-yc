# Merge Stockwatcher with CRSP

```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
pd.set_option('display.max_columns', None)
```

```python
sw_house = pd.read_csv("../data/house_all_transactions_fixed.csv")
sw_senate = pd.read_csv("../data/senate_all_transactions.csv")
sw_senate["transaction_date"] = list(map(lambda x: datetime.strptime(x, "%m/%d/%Y").strftime("%Y-%m-%d"), sw_senate["transaction_date"]))

crsp_house = pd.read_csv("../data/house_crsp.zip")
crsp_senate = pd.read_csv("../data/senate_crsp.zip")
```

## House

```python
sw_house.shape
```

```python
sw_crsp_house = sw_house.merge(crsp_house, how="left", left_on=["transaction_date", "ticker"], right_on=["dlycaldt", "ticker"])
sw_crsp_house
```

```python
sw_crsp_house.dropna(subset=["dlyprc"]).shape
```

```python
sw_crsp_house.dropna(subset=["dlyprc"]).shape[0] / sw_house.replace("--", np.nan).dropna(subset=["ticker"]).shape[0]
```

```python
# sw_crsp_house.to_csv("../data/sw_crsp_house.csv", index=False)
```

```python
compression_opts = dict(method='zip', archive_name='sw_crsp_house.csv')  
sw_crsp_house.to_csv('sw_crsp_house.zip', index=False,compression=compression_opts)
```

```python
%%bash
mv sw_crsp_house.zip ../data/sw_crsp_house.zip
```

## Senate

```python
sw_senate.shape
```

```python
sw_crsp_senate = sw_senate.merge(crsp_senate, how="left", left_on=["transaction_date", "ticker"], right_on=["dlycaldt", "ticker"])
sw_crsp_senate
```

```python
sw_crsp_senate.dropna(subset=["dlyprc"]).shape
```

```python
sw_crsp_senate.dropna(subset=["dlyprc"]).shape[0] / sw_senate.replace("--", np.nan).dropna(subset=["ticker"]).shape[0]
```

```python
# sw_crsp_senate.to_csv("../data/sw_crsp_senate.csv", index=False)
```

```python
compression_opts = dict(method='zip', archive_name='sw_crsp_senate.csv')  
sw_crsp_senate.to_csv('sw_crsp_senate.zip', index=False,compression=compression_opts)
```

```python
%%bash
mv sw_crsp_senate.zip ../data/sw_crsp_senate.zip
```
