```python
def fetch(queries, crsp_output):
    db = wrds.Connection(wrds_username='yiyangc')
    
    for sql in tqdm(queries["sql"], desc="Fetching CRSP"):
        r = db.raw_sql(sql, date_cols=['dlycaldt'])
        crsp_output = pd.concat([crsp_output, r]).reset_index(drop=True)
    
    db.close()
    
    return crsp_output
```

```python
sw_crsp = fetch(queries_test, crsp_output)
sw_crsp
```

```python
sw_crsp.to_csv("../data/sw_crsp.csv", index=False)
```
